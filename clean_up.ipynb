```python
every = pd.DataFrame(columns=['Time', 'total_carb', 'sugar', 'protein', 'Glucose'])
for i in range(14):
    if (i != 2) and (i != 7) and (i != 13):
        data = pd.read_csv(f'/Users/kevinli/Documents/GitHub/dsc106_final_project/filtered data/number{i+1}_integrate.csv')
    else:
        continue
    data = data.rename(columns = {'Timestamp (YYYY-MM-DDThh:mm:ss)': 'Time', 'Glucose Value (mg/dL)' : 'Glucose'})
    data['Time'] = pd.to_datetime(data['Time'])
    time = data[data['total_carb'].notna()].groupby('Time').sum()[['total_carb', 'sugar', 'protein','Glucose']]
    for index, content in time.iterrows():
        time1 = pd.to_datetime(index)
        index2 = data[data['Time'] == time1].index[0] - 1
        start_glucose = data.iloc[index2]['Glucose']
        start_time = time1 - pd.Timedelta(minutes= 5)
        end_time = time1 + pd.Timedelta(hours = 1)
        end_glucose = data[(data['Time'] >= start_time) & (data['Time'] <= end_time)]['Glucose'].max()
        time.loc[index, 'Glucose'] = end_glucose - start_glucose
    time = time.reset_index()
    every = pd.concat([every, time])
```
